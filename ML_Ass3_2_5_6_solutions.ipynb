```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import re
from sklearn import svm
```

Task 5

```python
# train SVM
X1_train = np.array([[1, 1],[2, 2],[2, 0],[0, 0],[1, 0],[0, 1]])
Y1_train = np.array([1, 1, 1, 0, 0, 0])
model = svm.SVC(kernel='linear', C=3)
model.fit(X1_train, Y1_train)
```

```python
# plot points
c1 = 'green'
c2 = 'orange'
plot.scatter(X1_train[:, 0], X1_train[:, 1], c=[c1, c1, c1, c2, c2, c2])
temp = np.linspace(-1,2.5)
x, y = np.meshgrid(temp, temp)
d = np.vstack([x.ravel(), y.ravel()]).T
z = model.decision_function(d).reshape(x.shape)
plot.contour(x, y, z, colors='red', levels=[0],linestyles=['dotted'])
plot.show()
```

```python
# plot hyperplan
c1 = 'green'
c2 = 'orange'
plot.scatter(X1_train[:, 0], X1_train[:, 1], c=[c1, c1, c1, c2, c2, c2])
temp = np.linspace(-1,2.5)
x, y = np.meshgrid(temp, temp)
d = np.vstack([x.ravel(), y.ravel()]).T
z = model.decision_function(d).reshape(x.shape)
plot.contour(x, y, z, colors='red', levels=[-1, 0, 1],linestyles=['dashed', 'dotted', 'dashed'])
plot.show()
print("Support Vectors = ",model.support_vectors_)
```
